```python
col_list = list(df.columns)
for x in col_list:
    if df[x].dtypes == 'object':
        df[x].fillna(df[x].mode()[0], inplace=True)
    else:
        df[x].fillna(df[x].median(), inplace=True)
```

```python
df.isna().sum()
```

```python
for x in col_list:
    if df[x].dtypes == 'object':
        continue
    plt.boxplot(df[x])
    plt.xlabel(x)
    plt.ylabel('count')
    plt.show()
```

```python
for x in col_list:
    if df[x].dtypes == 'object' or x == 'Gross':
        continue
    Q1 = df[x].quantile(0.25)
    Q3 = df[x].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    df = df[(df[x] >= lower_bound) & (df[x] <= upper_bound)]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for x in col_list:
    if df[x].dtypes == 'object':
        df[x] = le.fit_transform(df[x])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,:12] # we can also use -1.
Y = df.iloc[:,-1]   
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.75)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
model_predict = lr.predict(x_test)
```

```python
from sklearn.metrics import *
mse = mean_squared_error(y_test, model_predict)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
r2_accuracy = r2_score(y_test, model_predict)
r2_accuracy * 100 
```

```python
# Regression Plot 
sns.regplot(x = model_predict, y = y_test)
plt.title('Regression Plot')
plt.xlabel('Predictions')
plt.ylabel('Actual values')
plt.show()
```
