# 1. Check if a number is prime 

## Time: O(√n)
## Space: O(1)

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

# 2. Print all prime numbers from 1 to N
## Time: O(n√n)
## Space: O(1)

```python
def primes_upto_n(n):
    for num in range(2, n + 1):
        if is_prime(num):
            print(num, end=" ")
```

# 3. Factorial of a number
## Time: O(n)
## Space: O(1)

```python
def factorial(n):
    res = 1
    for i in range(2, n + 1):
        res *= i
    return res
```

# 4. Fibonacci series up to N terms
## Time: O(n)
## Space: O(1)

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b
```

# 5. Check palindrome number 
## Time: O(d)
## Space: O(d)

```python
def is_palindrome(n):
    return str(n) == str(n)[::-1]
```

# 6. Reverse a string
## Time: O(n)
## Space: O(n)

```python
def reverse_string(s):
    return s[::-1]
```

# 7. Sum of digits
## Time: O(d)
## Space: O(1)

```python
def sum_of_digits(n):
    s = 0
    while n:
        s += n % 10
        n //= 10
    return s
```

# 8. Count digits
## Time: O(d)
## Space: O(d)

```python
def count_digits(n):
    return len(str(n))
```

# 9. Largest of numbers
## Time: O(n)
## Space: O(1)


```python
def largest(nums):
    return max(nums)
```

# 10. Armstrong number
## Time: O(d)
## Space: O(d)

```python
def is_armstrong(n):
    digits = str(n)
    power = len(digits)
    return n == sum(int(d)**power for d in digits)
```

# 11. Pattern printing
## Time: O(n²)
## Space: O(1)

```python
def pattern(n):
    for i in range(1, n + 1):
        print("*" * i)
```
