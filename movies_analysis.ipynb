```python
df["YEAR"] = df["YEAR"].str.extract("(\d{4})")
df["YEAR"] = pd.to_numeric(df["YEAR"], errors="coerce")
```

```python
df["VOTES"] = df["VOTES"].str.replace(",", "")
df["VOTES"] = pd.to_numeric(df["VOTES"], errors="coerce")
```

```python
df["RATING"] = pd.to_numeric(df["RATING"], errors="coerce")
df["RunTime"] = pd.to_numeric(df["RunTime"], errors="coerce")
```

```python
df = df.dropna(subset=["RATING", "VOTES"])
```

```python
df.drop_duplicates()
```

# 1. Find the movies with the highest profit

```python
top_movies = df[["MOVIES","RATING"]].sort_values(by="RATING", ascending=False)
print(top_movies.head(10))
```

# 2. Find IMDB rank with top 100

```python
genres = df["GENRE"].str.split(",").explode().str.strip()
popular_genres = genres.value_counts()
print(popular_genres.head(10))
```

# 3. Find the best director

```python
popular_movies = df[["MOVIES","VOTES"]].sort_values(by="VOTES", ascending=False)
print(popular_movies.head(10))
```

# 4. Find the popular genres

```python
genre_rating = df.groupby("GENRE")["RATING"].mean().sort_values(ascending=False)
print(genre_rating.head(10))
```

# 5. Find the critic favorite and audience favorite actor

```python
df["Lead_Actor"] = df["STARS"].str.split(",").str[0]
actor_votes = df.groupby("Lead_Actor")["VOTES"].mean().sort_values(ascending=False)
print(actor_votes.head(10))
```
