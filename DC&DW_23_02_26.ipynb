```python
# Convert data types safely
numeric_columns = ['year', 'month', 'date_of_month', 'day_of_week', 'births']
for col in numeric_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Handle missing numeric values using median
for col in numeric_columns:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())
```

```python
# Create proper datetime column using correct parameter names
if all(col in df.columns for col in ['year', 'month', 'date_of_month']):
    df['full_date'] = pd.to_datetime({
        'year': df['year'],
        'month': df['month'],
        'day': df['date_of_month']
    }, errors='coerce')
```

```python
# Add weekday name and weekend flag
if 'full_date' in df.columns:
    df['weekday_name'] = df['full_date'].dt.day_name()
    df['is_weekend'] = df['weekday_name'].isin(['Saturday', 'Sunday']).astype(int)
```
