```python
# Remove duplicates
df = df.drop_duplicates(subset=["movie_title"])
```

```python
df["movie_title"] = df["movie_title"].str.strip()
```

```python
df = df.dropna(subset=["gross","budget","imdb_score"])
```

```python
df.isna().sum().sum()
```

```python
df = df.dropna(subset=["gross", "budget"])
```

```python
df = df.dropna(subset=["imdb_score"])
```

```python
df["director_name"] = df["director_name"].fillna("Unknown")
df["actor_1_name"] = df["actor_1_name"].fillna("Unknown")
df["actor_2_name"] = df["actor_2_name"].fillna("Unknown")
df["actor_3_name"] = df["actor_3_name"].fillna("Unknown")
```

```python
df["genres"] = df["genres"].fillna("Unknown")
```

```python
df.isnull().sum().sum()
```

```python
df
```

```python
# Create profit column
df["profit"] = df["gross"] - df["budget"]
```

# 1. Find the movies with the highest profit

```python
top_profit = df.sort_values("profit", ascending=False)[
    ["movie_title","profit"]
].head(10)

print(top_profit)
```

# 2. Top 100 Movies by IMDb Rank

```python
top100 = df.sort_values("imdb_score", ascending=False)[["movie_title","imdb_score"]].head(100)
print(top100)
```

# 3. Best Director

```python
best_director = df.groupby("director_name")["imdb_score"].mean().sort_values(ascending=False)
print(best_director.head(10))
```

# 4. Most Popular Genres

```python
popular_genres = df["genres"].str.split("|").explode().value_counts()
print(popular_genres)
```

# 5. Critic Favourite Actor

```python
critic_actor = df.melt(id_vars=["num_critic_for_reviews"],value_vars=["actor_1_name","actor_2_name","actor_3_name"],
value_name="actor")
critic_fav = critic_actor.groupby("actor")["num_critic_for_reviews"].mean().sort_values(ascending=False)
print(critic_fav.head(10))
```
