## 🔹 1. What is Hypothesis Testing?

Hypothesis Testing is a **statistical method** used to make decisions about a population using sample data.

👉 Core Idea:

* Assume something (hypothesis)
* Test using data
* Make a decision

---

## 🔹 2. Key Terminologies

### Null Hypothesis (H₀)

* Default assumption
* No effect / no difference

### Alternative Hypothesis (H₁)

* Opposes H₀

Types:

* Two-tailed → μ ≠ value
* Left-tailed → μ < value
* Right-tailed → μ > value

### Significance Level (α)

* Threshold for decision
* Common: 0.05

### p-value

* Probability of observed result under H₀

**Decision Rule:**

* p ≤ α → Reject H₀
* p > α → Fail to reject H₀

### Test Statistic

* Value used to test (Z, t, χ², F)

---

## 🔹 3. Errors

| Type    | Meaning                          |
| ------- | -------------------------------- |
| Type I  | Reject true H₀ (False Positive)  |
| Type II | Accept false H₀ (False Negative) |

---

## 🔹 4. Steps

1. Define H₀ and H₁
2. Choose α
3. Select test
4. Compute statistic
5. Find p-value
6. Make decision

---

## 🔹 5. Types of Tests

### Z-Test

* Large sample (n ≥ 30)
* Known variance

### T-Test

* Small sample
* Unknown variance

Types:

* One-sample
* Two-sample
* Paired

### Chi-Square Test

* Categorical data

### ANOVA

* Compare 3+ groups

---

## 🔹 6. One vs Two Tailed

| Test       | Use               |
| ---------- | ----------------- |
| One-tailed | Direction matters |
| Two-tailed | Any difference    |

---

## 🔹 7. Confidence Interval

* Range of expected values
* If CI contains null value → Do not reject H₀

---

## 🔹 8. Example

Claim: μ = 50

* H₀: μ = 50
* H₁: μ ≠ 50

If p = 0.03 and α = 0.05
→ Reject H₀

---

## 🔹 9. When to Use What

| Scenario     | Test       |
| ------------ | ---------- |
| Large sample | Z-test     |
| Small sample | T-test     |
| Categorical  | Chi-square |
| 3+ groups    | ANOVA      |

---

## 🔹 10. Assumptions

* Random data
* Independent samples
* Normal distribution (parametric)

---

## 🔹 11. Parametric vs Non-Parametric

| Parametric          | Non-Parametric        |
| ------------------- | --------------------- |
| Normal distribution | No assumption         |
| Example: t-test     | Example: Mann-Whitney |

---

## 🔹 12. Core Intuition

* You do NOT prove H₁
* You only reject or fail to reject H₀
* Lower p-value = stronger evidence

---

# 🚀 Things You MUST Remember (High-Impact Section)

## 🔥 Decision Logic

* p ≤ α → Reject H₀
* p > α → Fail to reject H₀

## 🔥 Meaning of p-value

* Not probability that H₀ is true
* It is probability of data assuming H₀ is true

## 🔥 Errors

* Type I → False Positive
* Type II → False Negative

## 🔥 Test Selection

* n ≥ 30 → Z-test
* n < 30 → T-test
* Categories → Chi-square
* Multiple groups → ANOVA

## 🔥 Hypothesis Structure

* H₀ → equality (=)
* H₁ → inequality (≠, <, >)

## 🔥 Confidence Insight

* CI includes null → No rejection

## 🔥 Golden Rule

* Hypothesis testing = decision under uncertainty

---

## 🔹 Python Example

```python
from scipy import stats

sample = [52, 55, 49, 50, 51, 53, 54]

t_stat, p_value = stats.ttest_1samp(sample, 50)

print(t_stat, p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to reject H0")
```
