# 1. What is OOP?
Object-Oriented Programming (OOP) is a programming paradigm where we model real-world entities as objects that contain:

`Data (attributes/variables)`

`Behavior (methods/functions)`

Instead of writing scattered functions, OOP helps you structure code like real-world systems.

# 🧠 OOP — Deep Conceptual Breakdown (Clear + Strong Mental Models)

---

# 🔐 1. Encapsulation (Control + Protection)

## 📌 Core Idea
Encapsulation is about **hiding internal data** and **controlling how it is accessed or modified**.

👉 Think like:  
You don’t directly touch a bank’s database. You use ATM/API.

---

## 🟢 Simple Understanding
- Wrap data + methods together  
- Restrict direct access using private variables (`__var`)  

**Why?**  
To prevent misuse and enforce rules.

---

## 🔴 Advanced Understanding (Industry Thinking)
Encapsulation is not just hiding — it’s about:

- Validation layers  
- Controlled state changes  
- Data integrity enforcement  

👉 Example mindset:  
“User should NOT directly modify balance → must go through rules like deposit/withdraw”

---

## ⚡ Real-world Mapping

| Concept       | Example              |
|--------------|--------------------|
| Private data | Bank balance        |
| Public method| deposit(), withdraw() |
| Control      | No negative transactions |

```python
# ================= ENCAPSULATION =================

# Basic Example
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

# Advanced Example
class SecureBankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Invalid transaction")

    def get_balance(self):
        return self.__balance
```


# 🧬 2. Inheritance (Reuse + Hierarchy)

## 📌 Core Idea
Inheritance allows one class to **reuse properties and behavior** of another.

---

## 🟢 Simple Understanding
- Child class inherits from parent  
- Avoid rewriting code  

👉 “Dog is an Animal”

---

## 🔴 Advanced Understanding (Industry Thinking)
Inheritance builds:

- Hierarchies  
- Extensibility  
- Code reuse at scale  

**BUT ⚠️**  
Bad inheritance = tight coupling  

👉 Modern systems often prefer **composition over inheritance**.

---

## ⚡ Real-world Mapping

| Parent   | Child   |
|----------|--------|
| Employee | Manager |
| Vehicle  | Car     |

```python
# ================= INHERITANCE =================

# Basic Example
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

# Advanced Example
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def get_details(self):
        return f"{self.name} earns {self.salary}"

class Manager(Employee):
    def __init__(self, name, salary, team_size):
        super().__init__(name, salary)
        self.team_size = team_size

    def get_details(self):
        return f"{self.name} manages {self.team_size} people and earns {self.salary}"
```


# 🎭 3. Polymorphism (Flexibility + Interchangeability)

## 📌 Core Idea
**Same interface → different behavior**

---

## 🟢 Simple Understanding
Different objects respond differently to the same method.

👉 `speak()` → Dog barks, Cat meows

---

## 🔴 Advanced Understanding (Industry Thinking)
Polymorphism enables:

- Plug-and-play architecture  
- Loose coupling  
- Scalable systems  

👉 You can swap components without breaking system  

**Examples:**
- Payment system → UPI / Card / PayPal  
- ML models → interchangeable pipelines  

---

## ⚡ Real-world Mapping

| Interface   | Variants           |
|------------|-------------------|
| Payment    | UPI, Card, Wallet |
| Notification | Email, SMS, Push |

```python
# ================= POLYMORPHISM =================

# Basic Example
class Cat:
    def speak(self):
        print("Meow")

class Dog:
    def speak(self):
        print("Bark")

# Advanced Example
class Payment:
    def pay(self, amount):
        raise NotImplementedError

class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid {amount} via Credit Card")

class UPI(Payment):
    def pay(self, amount):
        print(f"Paid {amount} via UPI")

class PayPal(Payment):
    def pay(self, amount):
        print(f"Paid {amount} via PayPal")
```

# 🧊 4. Abstraction (Simplification + Hiding Complexity)

## 📌 Core Idea
Expose only **what is necessary**, hide **how it works internally**

---

## 🟢 Simple Understanding
User clicks button → doesn’t know backend logic

---

## 🔴 Advanced Understanding (Industry Thinking)
Abstraction is:

- Interface design  
- System boundaries  
- Decoupling implementation from usage  

👉 You define **what should happen**, not **how**

---

## ⚡ Real-world Mapping

| Interface     | Hidden Logic        |
|--------------|-------------------|
| Database API | SQL execution      |
| Car steering | Engine mechanics   |

---

# 🧩 Final Mental Model (Critical)

| Concept       | Purpose           | Industry Value        |
|--------------|------------------|----------------------|
| Encapsulation | Protect data     | Security + validation |
| Inheritance   | Reuse code       | Faster development    |
| Polymorphism  | Flexibility      | Scalable systems     |
| Abstraction   | Hide complexity  | Clean architecture   |  

```python
# ================= ABSTRACTION =================

# Basic Example
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

# Advanced Example
class Database(ABC):
    @abstractmethod
    def connect(self):
        pass

    @abstractmethod
    def execute(self, query):
        pass

class MySQLDatabase(Database):
    def connect(self):
        print("Connected to MySQL")

    def execute(self, query):
        print(f"Executing {query} on MySQL")

class MongoDB(Database):
    def connect(self):
        print("Connected to MongoDB")

    def execute(self, query):
        print(f"Executing {query} on MongoDB")
```
