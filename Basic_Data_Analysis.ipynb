## 1. Objective

Perform basic data manipulation, visualization, and analysis on the given dataset to extract meaningful insights.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv(r'Datasets/basketball_heights.csv')
df
```

## 3. Data Understanding

### 🔹 Explanation

* `info()` checks data types and missing values
* `describe()` gives statistical summary

### 🔹 Insights

* Missing values exist in `region_1` and `region_3`
* `region_2` is complete

```python
df.info()
df.describe()
```

## 4. Handling Missing Values

### 🔹 Explanation

* Replaced missing values with column mean
* Ensures dataset is usable for analysis

### 🔹 Insights

* Prevents bias from dropping rows
* Maintains distribution consistency


```python
df_filled = df.fillna(df.mean())
```

## 5. Data Visualization

### 🔹 Explanation

* Histogram shows distribution of values in each region

### 🔹 Insights

* All regions show similar distribution
* Slight variation in spread

```python
df_filled.hist(figsize=(8,5))
plt.show()
```

## 6. Correlation Analysis

### 🔹 Explanation

* Measures relationship between variables

### 🔹 Insights

* Regions are moderately correlated
* Indicates possible similar behavior patterns

```python
df_filled.corr()
```
