# 5. Program to find factorial of a number

```python
# Answer - 
n = int(input())
fact = 1

for i in range(1, n + 1):
    fact *= i 

print(fact)
```

# 7. Program to check if a number is prime

```python
# Answer -
n = int(input())
if n <= 1:   # Numbers less than or equal to 1 cannot be prime by definition.
    print("Not Prime", n)
else:
    for i in range(2, n):   # Now we test whether n can be divided by any number between 2 and n-1.
        if n % i == 0:
            print("It's not prime:", n)
            break
    else:
        print("It's prime:", n) 
```

# 8. Difference between append() and extend()

```python
# Answer - 
# append(): Add one item.
# extend(): Merge another iterable into the list 
# Use append() for adding individual items.
# Use extend() for combining lists or adding multiple elements efficiently. 
a = [1, 2, 3]

a.append([4, 5])
print(a)        # [1, 2, 3, [4, 5]]

a = [1, 2, 3]
a.extend([4, 5])
print(a)        # [1, 2, 3, 4, 5] 
```

# 9. Difference between break and continue

```python
# break() - exits the loop permanently
# continue() - skips the current iteration 

for i in range (1, 5):
    if i == 3:
        break
    print("Using break statement:", i)
    
for j in range (1, 5):
    if j == 3:
        continue
    print("Using Continue statement:", j)
```
