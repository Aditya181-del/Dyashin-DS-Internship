```python
df.drop(columns=['index'], inplace=True)
```

```python
df
```

```python
df.replace('?', np.nan, inplace=True)
df
```

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Fill numeric missing values with median
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].apply(lambda x: x.fillna(x.median()))
```

```python
categorical_cols = df.select_dtypes(exclude=np.number).columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
if 'price' in df.columns:
    df['price_category'] = pd.qcut(df['price'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
print("\nCleaned Dataset Shape:", df.shape)
print("\nMissing Values After Cleaning:")
display(df.isnull().sum())
```

```python
avg_mileage = df['average-mileage'].mean()
print("Average Mileage:", avg_mileage)
```

```python
mileage_company = df.groupby('company')['average-mileage'].mean()
print(mileage_company)
```

```python
mileage_body = df.groupby('body-style')['average-mileage'].mean()
print(mileage_body) 
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='company', y='average-mileage', data=df)
plt.xticks(rotation=45)
plt.title("Average Mileage by Company")
plt.show()
```

```python
high_mileage = df[df['average-mileage'] > 30]
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='company', y='average-mileage', data=high_mileage)
plt.title("Cars with Mileage Greater than 30")
plt.xlabel("Company")
plt.ylabel("Average Mileage")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_hp = df.groupby('company')['horsepower'].mean()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=avg_hp.index, y=avg_hp.values)
plt.title("Average Horsepower by Company")
plt.xlabel("Company")
plt.ylabel("Average Horsepower")
plt.xticks(rotation=45)
plt.show()
```
